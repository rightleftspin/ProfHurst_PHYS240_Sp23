The Problem:
Solve the equation: a - v^2 + x + 5 = 0  using Leap-Frog Method

Initial conditions:
x_0 = 1
v_0 = 0

Solve from t = 0 to t = 50 seconds in .001 second time-steps
Plot results

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 1 #list initial/given values first
v0 = 0
t0 = 0
tau = .001 #timestep
nStep = int(50/tau + 1) #number of iterations to get from t=0 to t=50

xplot = np.zeros(nStep) #create all the arrays we'll need
vplot = np.zeros(nStep)
tplot = np.empty(nStep)
accel = np.zeros(nStep)
xplot[0] = x0 #set initial x value, v=0 already
accel[0] = -6 #solved with pen & paper

for i in range (nStep-1): #leapfrog method
    tplot[i] = i*tau
    xplot[i+1] = xplot[i]+tau*vplot[i]+0.5*accel[i]*tau**2
    accel[i+1] = vplot[i]**2-xplot[i+1]-5
    vplot[i+1] = vplot[i] + 0.5 * tau * (accel[i+1] + accel [i-1])
    
fig = plt.plot() #had problems using matplotlib; I find their website difficult to navigate
plt.plot(tplot,xplot,'.',label='Position')
plt.xlabel('Time (sec)')
plt.ylabel('Position (arb. units)')
plt.title('Position over Time')
plt.grid(True) 
plt.show()     
```

```python
fig = plt.plot() #had problems using matplotlib; I find their website difficult to navigate
plt.plot(tplot,vplot,'.',color='green',label='Velocity')
plt.xlabel('Time (sec)')
plt.ylabel('Velcoity (arb./sec)')
plt.title('Velocity over Time')
plt.grid(True) 
plt.show()     
```
