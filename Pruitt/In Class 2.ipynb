```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def partialSum(x, N):
    S = 0
    for i in range(N):
        S += x**i/np.math.factorial(i)
    return S

```

```python
testVals = np.array([-10, -2, 2, 10])
for val in testVals:
    print(np.array(partialSum(val, 60)))
```

```python
def frac_error(x,N):
    output = np.abs(partialSum(x,N) - np.exp(x))/np.exp(x)
    return output
```

```python
for val in testVals:
    print(frac_error(val, 60))
```

```python
x0 = 10
yvals = []
Nvals = np.arange(0,60,1)

for val in Nvals:
    yvals.append(frac_error(x0, val))
plt.figure()
plt.loglog(Nvals, yvals, 'k.')
```

```python
x1 = 2
yvals = []
Nvals = np.arange(0,60,1)

for val in Nvals:
    yvals.append(frac_error(x1, val))
plt.figure()
plt.loglog(Nvals, yvals, 'k.')
```
